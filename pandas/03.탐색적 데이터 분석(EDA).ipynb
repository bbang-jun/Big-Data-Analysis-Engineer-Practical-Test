```python
import pandas as pd

df = pd.DataFrame({
    "메뉴" : ["아메리카노", "카페라떼", "카페모카", "카푸치노", "에스프레스", "밀크티", "녹차"],
    "가격" : [4500, 5000, 5500, 5000, 4000, 5900, 5300],
    "칼로리" : [10, 110, 250, 110, 20, 210, 0]
})

df
```

```python
# 상위 N개의 행 반환
df.head(3)
```

```python
# 하위 N개의 행 반환
df.tail(4)
```

```python
# 샘플링(데이터프레임에서 임의로 샘플링해 N개의 행 반환)
df.sample(3)
```

```python
# 데이터프레임의 크기 확인
df.shape
```

```python
# 데이터프레임 안 컬럼의 자료형 확인
df.info()
```

```python
# 상관 관계 확인
# 값 범위: -1 ~ 1
# 1 → 완벽한 양의 상관 (같이 증가)
# -1 → 완벽한 음의 상관 (하나는 증가, 하나는 감소)
# 0 → 관계 없음
# numeric_only: 문자형 컬럼으로 인한 convert 에러 방지
df.corr(numeric_only=True)
```

```python
df_car = pd.DataFrame({
"car":['Sedan', 'SUV', 'Sedan', 'SUV','SUV','SUV','Sedan', 'Sedan', 'Sedan', 'Sedan', 'Sedan'],
"size": ['S','M','S','S','M', 'M','L','S','S', 'M','S']
} )

df_car
```

```python
# unique(): 시리즈에서 고유한 값
df_car['car'].unique()

# nunique(): 시리즈에서 고유한 값 개수
df_car['car'].nunique()
```

```python
# nunique(): number of unique로, 고유한 값의 개수
df_car.nunique()
```

```python
# value_counts(): unique()와 nuique()의 결과 내용을 한 번에 확인
print(df_car['car'].value_counts())
print()
print(df_car['size'].value_counts())
```

```python
# describe(): 수치형 데이터에 대한 기술통계량
# count: 값이 있는 데이터 수, mean: 평균, std: 표준편차, min: 최솟값, 00%: 백분위수에서 00%, max: 최댓값
df.describe()
```

```python
# describe(include="O"): 범주형(object 자료형) 데이터에 대한 기술통계량
# count: 값이 있는 데이터 수, unique: 고유한 데이터의 수(종류), top: 가장 많이 나오는 값(최빈값), freq: 가장 많이 나오는 값의 빈도 수
df_car.describe(include="O")
```
