```python
# 런타임 > 세션 관리에서 기존 세션들을 끊어야 새로운 런타임 연결이 가능함.
# 좌측 파일 탭에서 업로드, 새로고침 버튼 오른쪽의 마운트 진행
import pandas as pd
%cd /content/drive/MyDrive/빅데이터분석기사_실기/Big-Data-Analysis-Engineer-Practical-Test/pandas
```

```python
df = pd.read_csv("cafe3.csv")
df.head(3)
```

```python
# replace(): 특정 값을 찾아 대체하는 함수. 변경 전 값과 변경 후 값을 입력함.
df.replace('아메리카노', '룽고', inplace=True)
df.replace('녹차',  '그린티', inplace=True)
df
```

```python
# 딕셔너리 기반 replace
change = {'룽고':'아메리카노', '그린티':'녹차'}
df.replace(change, inplace=True)
df
```

```python
df.loc[6, '원산지'] = '대한민국'
df.tail(3)
```

```python
df.loc[1:2, '이벤트가'] = 1000
df.head()
```

```python
df.to_csv("cafe4.csv", index=False)
```

```python
# map(): replace()에 비해 유연성이 뛰어남. map()에서 없는 값이면 nan이 되지만, replace()에서 없는 값이면 기존 값이 유지됨.
df = pd.DataFrame({'만족도' : ['만족', '보통', '불만', '보통', '만족']})

mapping = {'만족' : 3, '보통' : 2, '불만' : 1}
df['만족도_숫자'] = df['만족도'].map(mapping)
df
```
