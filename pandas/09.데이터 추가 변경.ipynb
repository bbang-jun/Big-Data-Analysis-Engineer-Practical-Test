```python
df = pd.read_csv('cafe.csv')
df.head(2)
```

```python
# 넘파이를 활용하여 결측지(NaN)가 있는 데이터 생성
import numpy as np
df['원산지'] = np.nan
df
```

```python
df.loc[0, "원산지"] = '콜롬비아'
df.loc[2:3, "원산지"] = '과테말라'
df.head(4)
```

```python
df.loc['시즌'] = ['크리스마스라떼', 6000, 300, '한국']
df
```

```python
# 인덱스명 7로 딕셔너리 형태로 행 추가. 원산지 컬럼은 지정하지 않았기 때문에 결측치인 NaN으로 채워짐.
df.loc[7] = {"메뉴" : "딴짓커피", "가격" : 2000, "칼로리" : 20}
df
```

```python
df = df.drop("시즌", axis=0)
df
```

```python
df.to_csv('cafe2.csv', index=False)
```
