```python
# drop(): 행 혹은 열 삭제
# axis=0: 행, axis=1: 열
# inplace 파라미터를 사용해야 삭제된 결과값이 저장됨. 혹은 inplace대신 '='으로 저장
df.drop(1, axis=0, inplace=True)
df.head(3)
```
