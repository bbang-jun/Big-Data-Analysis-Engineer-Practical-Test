```python
# 런타임 > 세션 관리에서 기존 세션들을 끊어야 새로운 런타임 연결이 가능함.
# 좌측 파일 탭에서 업로드, 새로고침 버튼 오른쪽의 마운트 진행
import pandas as pd
%cd /content/drive/MyDrive/빅데이터분석기사_실기/Big-Data-Analysis-Engineer-Practical-Test/pandas
```

```python
df = pd.read_csv("cafe4.csv")
df.head(2)
```

```python
# groupby(): 그룹핑을 수행하는 함수이며, 일반적으로 대괄호는 2개 이상일 때 사용함.
df.groupby('원산지').mean(numeric_only=True)
```

```python
df.groupby(['원산지', '칼로리']).mean(numeric_only=True)
```

```python
# agg(): 여러 개의 컬럼에 대해 다양한 집계 연산 동시 수행
df.groupby(['원산지', '메뉴']).agg(['mean', 'sum'])
```

```python
# 그룹핑하여 인덱스가 사라짐.(reset_index() 필요)
df.groupby(['원산지', '칼로리']).mean(numeric_only=True)
```

```python
# reset_index(): 그룹핑하여 인덱스가 사라진 상태에서 reset_index()를 사용하여 인덱스 생성
df.groupby(['원산지', '칼로리']).mean(numeric_only=True).reset_index()
```

```python
df2 = pd.DataFrame({
    '과일' : ['딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리','딸기', '블루베리'],
    '가격' : [1000, None, 1500, None, 2000, 2500, None, 1800]

})

df2
```

```python
# 과일의 그룹 수만큼 행이 줄어듬.
df2.groupby('과일').mean()
```

```python
# transform(): 인자의 연산으로 집계는 하되, 원래 행마다 값을 붙여줌.
# 그룹 바이 후, 결측치를 계산에서 제외한 평균을 각 행마다 적어줌.(결측치가 있는 행에도 적어줌)
price = df2.groupby('과일')['가격'].transform("mean")
price
```

```python
# fillna로 결측값만 채움
df2['가격'] = df2['가격'].fillna(price)
df2
```

```python
df3 = pd.DataFrame({
    '과일' : ['딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리','딸기', '블루베리'],
    '등급' : ['B', 'B', 'A', 'A', 'A', 'A', 'B', 'B'],
    '가격' : [1000, None, 1500, None, 2000, 2500, None, 1800]
})

df3
```

```python
# 예제3
price = df3.groupby(['과일', '등급'])['가격'].transform("mean")
price
```
