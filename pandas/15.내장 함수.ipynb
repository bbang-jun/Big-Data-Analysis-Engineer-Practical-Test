```python
# 런타임 > 세션 관리에서 기존 세션들을 끊어야 새로운 런타임 연결이 가능함.
# 좌측 파일 탭에서 업로드, 새로고침 버튼 오른쪽의 마운트 진행
import pandas as pd
%cd /content/drive/MyDrive/빅데이터분석기사_실기/Big-Data-Analysis-Engineer-Practical-Test/pandas
```

```python
df = pd.read_csv("cafe4.csv")
df.head(2)
```

```python
# 행의 수
print(len(df))
# 행의 수, 열의 수
print(df.shape[0], df.shape[1])
```

```python
cond = df['가격'] > 5000

# sum으로 조건에 맞는 개수 구하기
print(sum(cond))
# len으로 조건에 맞는 개수 구하기
print(len(df[cond]))
```

```python
df.sum(numeric_only=True)
```

```python
# 행과 열을 서로 변경
df.T
```

```python
# 열별 합계
df.sum(numeric_only=True) # df.sum(axis=0, numeric_only=True)
# 행별 합계
# axis=1: 행, axis=0: 열
# drop과 축설정이 반대
df.sum(axis=1, numeric_only=True)
```

```python
# max(), min(), mean(), median(), sum(), std(), var()
print("최댓값: ", df['가격'].max())
print("최솟값: ", df['가격'].min())
print("평균값: ", df['가격'].mean())
print("중앙값: ", df['가격'].median())
print("합계: ", df['가격'].sum())
print("표준편차: ", df['가격'].std())
print("분산: ", df['가격'].var())
```

```python
# quantile(): 분위 수 구하기
print("분위수 25% 값:", df['가격'].quantile(.25))
print("분위수 75% 값:", df['가격'].quantile(0.75))
```

```python
# 1사분위수보다 작은 가격의 데이터 출력
cond = df['가격'].quantile(.25) > df['가격']
df[cond]
```

```python
# 3사분위수보다 큰 가격의 데이터 출력
third_quantile = df['가격'].quantile(0.75)
cond = df['가격'] > third_quantile
df[cond]
```

```python
# mode(): 최빈값
# 최빈값은 여러 개일 수도 있으므로, 시리즈 형태로 반환됨. 따라서 인덱스로 원하는 값을 꺼내와야 함.
df['원산지'].mode()[0]
```

```python
# idxmax(): 열에서 최댓값을 갖는 인덱스 반환
# idxmin(): 열에서 최솟값을 갖는 인덱스 반환
print('가격이 가장 큰 인덱스와 가격:', df['가격'].idxmax(), df['가격'].max())
print('가격이 가장 작은 인덱스와 가격:', df['가격'].idxmin(), df['가격'].min())

max_idx = df['가격'].idxmax()
df.loc[max_idx]
```

```python
# 효율적
print(df.loc[max_idx, '메뉴'])
# 비효율적
print(df.loc[max_idx]['메뉴'])
```

```python
# nlargest(): 데이터프레임의 특정 컬럼 중에 큰 값 순으로 n개 반환
df.nlargest(3, '가격')
```

```python
# nsmallest(): 데이터프레임의 특정 컬럼 중에 작은 값 순으로 n개 반환
df.nsmallest(2, '가격')
```

```python
# apply(): 적용할 함수명을 인자로 줄 수 있음
def cal(x):
  if x >= 100:
    return "No"
  else:
    return "Yes"

df['먹어도 될까요'] = df['칼로리'].apply(cal)
df
```
