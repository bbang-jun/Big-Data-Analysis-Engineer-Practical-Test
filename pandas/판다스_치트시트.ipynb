```python
# 14.문자열


# 단어의 일부분은 변경할 수 없음
df['A'] = df['A'].replace('분석', '시각화')
# str.replace(): str 접근자를 활용하여 전체 문자열 중 일부만 변경
df['A'] = df['A'].str.replace('분석', '시각화')

# int형 정수에 str.replace()를 사용하면 에러 발생
df['B'] = df['B'].str.replace(20, 200)

# str.split(): 문자열 분리
# 문자열을 리스트로 나눠 시리즈 자료형으로 반환
# 0	[데이터, 시각화]
# 1	[기본, 학습서]
# 2	[퇴근, 후, 열공]
df['A'].str.split()

# 특정 행의 데이터 분리
# ['데이터', '시각화']
df['A'].str.split()[0]

# str.split().str[0]: 분리 후 첫 번째 단어
# A	B	C	D
# 0	데이터 시각화	100	ab cd	데이터
# 1	기본 학습서	20	AB CD	기본
# 2	퇴근 후 열공	30	ab cd	퇴근
df['D'] = df['A'].str.split().str[0]

# str.contains(): 특정 단어가 포함되어 있는지 확인
# isin()은 값을 찾고, contains()는 값을 찾거나 값의 일부를 찾음.
# isin()은 여러 개의 값을 찾을 수 있는 장점이 있어 필터링에서 유용하게 사용되지만, 1개 값만 찾더라도 .isin(['맛난버거'])와 같이 대괄호 필요
# str.contains()는 문장 속에서 특정 문자열을 찾을 수 있는 장점이 있으며, 리스트 형태가 아니므로 대괄호는 필요 없음.
df['기본포함유무'] = df['A'].str.contains('기본')


# 0    False
# 1    False
# 2    False
# Name: A, dtype: bool
# 0    False
# 1     True
# 2    False
# Name: A, dtype: bool
print(df['A'].isin(['기본']))
print(df['A'].str.contains('기본'))

# str.len(): 문자열 길이
df['문자길이'] = df['A'].str.len()

# 대문자 -> 소문자
df['C'] = df['C'].str.lower()

# 소문자 -> 대문자
df['C'] = df['C'].str.upper()

# 소문자로 변환 후 공백 제거
df['C'] = df['C'].str.lower()
df['C'] = df['C'].str.replace(" ", "")

df['C'].str.strip() # 양쪽 공백 제거
df['C'].str.lstrip() # 왼쪽 공백 제거
df['C'].str.rstrip() # 오른쪽 공백 제거
```

```python
# 15.내장 함수


# 열별 합계
df.sum(numeric_only=True) # df.sum(axis=0, numeric_only=True)
# 행별 합계 / axis=1: 행, axis=0: 열 / drop과 축설정이 반대
df.sum(axis=1, numeric_only=True)

# max(), min(), mean(), median(), sum(), std(), var()
print("최댓값: ", df['가격'].max())
print("최솟값: ", df['가격'].min())
print("평균값: ", df['가격'].mean())
print("중앙값: ", df['가격'].median())
print("합계: ", df['가격'].sum())
print("표준편차: ", df['가격'].std())
print("분산: ", df['가격'].var())

# quantile(): 분위 수 구하기
print("분위수 25% 값:", df['가격'].quantile(.25))
print("분위수 75% 값:", df['가격'].quantile(0.75))
# 1사분위수보다 작은 가격의 데이터 출력
cond = df['가격'].quantile(.25) > df['가격']
df[cond]
# 3사분위수보다 큰 가격의 데이터 출력
third_quantile = df['가격'].quantile(0.75)
cond = df['가격'] > third_quantile
df[cond]

# mode(): 최빈값
# 최빈값은 여러 개일 수도 있으므로, 시리즈 형태로 반환됨. 따라서 인덱스로 원하는 값을 꺼내와야 함.
df['원산지'].mode()[0]

# idxmax(): 열에서 최댓값을 갖는 인덱스 반환
# idxmin(): 열에서 최솟값을 갖는 인덱스 반환
print('가격이 가장 큰 인덱스와 가격:', df['가격'].idxmax(), df['가격'].max())
print('가격이 가장 작은 인덱스와 가격:', df['가격'].idxmin(), df['가격'].min())
max_idx = df['가격'].idxmax()
df.loc[max_idx]

# 효율적
print(df.loc[max_idx, '메뉴'])
# 비효율적
print(df.loc[max_idx]['메뉴'])

# nlargest(): 데이터프레임의 특정 컬럼 중에 큰 값 순으로 n개 반환
df.nlargest(3, '가격')
# nsmallest(): 데이터프레임의 특정 컬럼 중에 작은 값 순으로 n개 반환
df.nsmallest(2, '가격')

# apply(): 적용할 함수명을 인자로 줄 수 있음
def cal(x):
  if x >= 100:
    return "No"
  else:
    return "Yes"

df['먹어도 될까요'] = df['칼로리'].apply(cal)
df
```
