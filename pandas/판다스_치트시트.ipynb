```python
import pandas as pd
```

```python
# pd.Series()
# 시리즈 만들기

menu = pd.Series(['비빔밥', '김치찌개', '된장찌개'])
price = pd.Series([10000, 9000, 8000])
```

```python
# pd.DataFrame()
# 데이터 프레임 만들기

pd.DataFrame({
    "메뉴" : menu,
    "가격" : price
})

df = pd.DataFrame({
    "메뉴" : ["비빔밥", "김치찌개", "된장찌개"],
    "가격" : [10000, 9000, 8000],
    "원산지" : ["국내산", "국내산", "국내산"]
})
```

```python
# csv 파일로 저장하기
df.to_csv('temp.csv')

# csv 파일로 저장하기(인덱스 제외)
df.to_csv('temp.csv', index=False)


# csv 파일 불러오기
temp_df = pd.read_csv('temp.csv')

# csv 파일 불러오기(인덱스로 사용할 컬럼명 또는 열의 번호 지정)
pd.read_csv('temp.csv', index_col='컬럼명')
pd.read_csv('temp.csv', index_col=0)

# csv 파일 불러오기(불러올 컬럼명 또는 열의 번호 지정)
pd.read_csv('temp.csv', usecols=['메뉴', '가격'])
pd.read_csv('temp.csv', usecols=[0, 1])
```
