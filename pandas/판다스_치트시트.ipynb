```python
# 넘파이를 활용하여 결측지(NaN)가 있는 데이터 생성
import numpy as np
df['원산지'] = np.nan

# 인덱스명이 0인 행의 원산지 컬럼의 값을 콜롬비아로 변경
df.loc[0, "원산지"] = '콜롬비아'
# 인덱스명이 2~3인 행의 원산지 컬럼의 값을 과테말라로 변경
df.loc[2:3, "원산지"] = '과테말라'

# 인덱스명 7로 딕셔너리 형태로 행 추가. 원산지 컬럼은 지정하지 않았기 때문에 결측치인 NaN으로 채워짐.
df.loc[7] = {"메뉴" : "딴짓커피", "가격" : 2000, "칼로리" : 20}
```

```python
# sort_index(): 인덱스 기준 정렬
df.sort_index(ascending=False)

# sort_values(): 데이터 값 기준 정렬
df.sort_values('메뉴', ascending=False)

# 가격 내림차순, 메뉴 오름차순(= 가격 DESC, 메뉴 ASC)
df = df.sort_values(['가격', '메뉴'], ascending=[False, True])

# df의 인덱스명 순서가 변경되었으므로, 다시 0부터 7까지 새롭게 지정
df = df.reset_index(drop=True)
```

```python
# ~: NOT연산자
df[~cond]

# & 연산자: AND 연산이며, 가격이 5000원 이상이면서 칼로리가 100보다 큰 데이터 필터링
cond1 = (df['가격'] >= 5000)
cond2 = (df['칼로리'] > 100)
df[cond1 & cond2]

# | 연산자: OR 연산이며, 가격이 5000원 이상이거나 칼로리가 100보다 큰 데이터 필터링
cond1 = (df['가격'] >= 5000)
cond2 = (df['칼로리'] > 100)
df[cond1 | cond2]

# isin(): 주어진 값이 있는지 확인
cond = df['메뉴'].isin(['녹차'])
df[cond]

box = ['녹차', '카푸치노', '카페라떼']
cond = df['메뉴'].isin(box)
df[cond]
```
