```python
# & 연산자: AND 연산이며, 가격이 5000원 이상이면서 칼로리가 100보다 큰 데이터 필터링
cond1 = (df['가격'] >= 5000)
cond2 = (df['칼로리'] > 100)
df[cond1 & cond2]
```

```python
# | 연산자: OR 연산이며, 가격이 5000원 이상이거나 칼로리가 100보다 큰 데이터 필터링
cond1 = (df['가격'] >= 5000)
cond2 = (df['칼로리'] > 100)
df[cond1 | cond2]
```

```python
cond = (df['원산지'] == '과테말라')
df[cond]
```
