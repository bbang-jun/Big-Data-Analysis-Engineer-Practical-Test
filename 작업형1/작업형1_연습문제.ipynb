```python
# Section 01 필터링, 최솟값, 중앙값
# 1. 'f5' 컬럼이 0이 아닌 데이터(행)를 구하시오.
# 2. 앞에서 구한 데이터에 'views' 컬럼 결측치를 'views' 컬럼의 최솟값으로 채워주세요.
# 3. 그리고 'views' 컬럼의 중앙값을 계산해 정수로 구하시오.

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'f5' 컬럼이 0이 아닌 데이터(행)를 구하시오.
cond = df['f5'] != 0
df = df[cond]
df
```

```python
# 2. 앞에서 구한 데이터에 'views' 컬럼 결측치를 'views' 컬럼의 최솟값으로 채워주세요.
min = df['views'].min() # 42.0
df['views'] = df['views'].fillna(min)
# df[df['views'].isnull()] # 결측치가 하나도 없으므로, 최솟값으로 잘 채워진 것임.
df
```

```python
# 3. 그리고 'views' 컬럼의 중앙값을 계산해 정수로 구하시오.
mid = df['views'].median() # 5924.0
print(int(mid))
```

```python
# Section 02 카테고리, 인덱스, 문자열 슬라이싱
# 1. 'subscribed' 컬럼에서 가장 빈도수가 많은 날짜를 구하시오.
# 2. 앞에서 구한 날짜의 일(day) 값을 정수로 구하시오.

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 2. 앞에서 구한 날짜의 일(day) 값을 정수로 구하시오.
print(int(df.index[0][-2:])) # .index[]는 n번째 인덱스를 가져옴
```

```python
# Section 03 파생변수, 정렬, 인덱싱
# 1. 결측치가 있는 데이터(행)를 제거하시오.
# 2. 'views' 컬럼을 'f1' 컬럼으로 나눈 값을 새로운 컬럼으로 추가하시오.
# 3. 새로운 컬럼 값 중 가장 큰 값을 가진 행의 age를 정수로 구하시오.

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 결측치가 있는 데이터(행)를 제거하시오.
df = df.dropna()
df
```

```python
# 2. 'views' 컬럼을 'f1' 컬럼으로 나눈 값을 새로운 컬럼으로 추가하시오.
df['new'] = df['views'] / df['f1']
df
```

```python
# 3. 새로운 컬럼 값 중 가장 큰 값을 가진 행의 age를 정수로 구하시오.
df.sort_values('new', ascending = False)
print(int(df.iloc[0][1]))
```

```python
# Section 04 값 변경, 정렬, 합계
# 1. 'views' 컬럼의 결측 데이터를 0으로 대체하시오.
# 2. 'views' 컬럼에서 상위 10번째 값을 구하시오.
# 3. 'views' 컬럼에서 상위 10개의 값을 상위 10번째 값으로 대체하시오.
# 4. 'views' 컬럼의 전체 합을 정수로 구하시오.

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'views' 컬럼의 결측 데이터를 0으로 대체하시오.
# f1, f3, views 결측치 존재
# df.info()
df['views'] = df['views'].fillna(0)
```

```python
# 2. 'views' 컬럼에서 상위 10번째 값을 구하시오.
df = df.sort_values('views', ascending = False)
num = df.iloc[9]['views']
print(num)
```

```python
# 3. 'views' 컬럼에서 상위 10개의 값을 상위 10번째 값으로 대체하시오.
df.loc[:9, 'views'] = num
df
```

```python
# 4. 'views' 컬럼의 전체 합을 정수로 구하시오.
print(int(df['views'].sum()))
```

```python
# Section 05 문자열 슬라이싱, 파생변수, 평균값
# 1. 'f4' 컬럼 데이터에 'FJ'가 포함된 데이터를 찾으시오.
# 2. 찾은 데이터 중에서 'f2' 컬럼의 평균값을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'f4' 컬럼 데이터에 'FJ'가 포함된 데이터를 찾으시오.
df['new'] = df['f4'].str.contains('FJ')
df
```

```python
# 2. 찾은 데이터 중에서 'f2' 컬럼의 평균값을 구하시오. (반올림 후 소수 둘째 자리까지 계산)
cond = (df['new'] == True)

print(round(df[cond]['f2'].mean(), 2))
```

```python
# Section 06 필터링, 분산
# 1. 'f3' 컬럼이 gold이면서 'f2' 컬럼이 2인 데이터를 찾으시오.
# 2. 찾은 데이터에서 'f1' 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'f3' 컬럼이 gold이면서 'f2' 컬럼이 2인 데이터를 찾으시오.
cond = ((df['f3'] == 'gold') & (df['f2'] == 2))
df = df[cond]
df
```

```python
# 2. 찾은 데이터에서 'f1' 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)
print(round(df['f1'].var(), 2))
```

```python
# Section 07 값 변경(연산), 필터링 절댓값
# 1. 모든 나이(age)에 1을 더하시오.
# 2. 20대의 'views' 평균과 30대의 'views' 평균의 차이의 절댓값을 구하시오.(반올림 후 소수 둘째 자리까지 계산)

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 모든 나이(age)에 1을 더하시오.
df['age'] = df['age'] + 1
df
```

```python
# 2. 20대의 'views' 평균과 30대의 'views' 평균의 차이의 절댓값을 구하시오.(반올림 후 소수 둘째 자리까지 계산)
cond1 = (df['age'] >= 20) & (df['age'] < 30)
cond2 = (df['age'] >= 30) & (df['age'] < 40)
avg_20 = df[cond1]['views'].mean()
avg_30 = df[cond2]['views'].mean()
avg_abs = avg_20 - avg_30
print(round(abs(avg_abs), 2))
```

```python
# Section 08 시계열 데이터, 필터링, 데이터 개수
# 1. 'subscribed' 컬럼이 2024년 2월인 데이터를 찾으시오.
# 2. 위에서 찾은 데이터 중 'f3' 컬럼이 gold인 데이터의 개수를 구하시오.

# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'subscribed' 컬럼이 2024년 2월인 데이터를 찾으시오.
df['subscribed'] = pd.to_datetime(df['subscribed'], format = '%Y-%m-%d')
cond = (df['subscribed'].dt.year == 2024) & (df['subscribed'].dt.month == 2)
df = df[cond]
df
```

```python
# 2. 위에서 찾은 데이터 중 'f3' 컬럼이 gold인 데이터의 개수를 구하시오.
cond = (df['f3'] == 'gold')
print(len(df[cond]))
```
