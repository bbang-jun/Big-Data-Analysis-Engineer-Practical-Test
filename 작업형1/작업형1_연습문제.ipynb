```python
# Section 01 필터링, 최솟값, 중앙값
1. 'f5' 컬럼이 0이 아닌 데이터(행)를 구하시오.
2. 앞에서 구한 데이터에 'views' 컬럼 결측치를 'views' 컬럼의 최솟값으로 채워주세요.
3. 그리고 'views' 컬럼의 중앙값을 계산해 정수로 구하시오.
```

```python
# 기본 코드
import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# 1. 'f5' 컬럼이 0이 아닌 데이터(행)를 구하시오.
cond = df['f5'] != 0
df = df[cond]
df
```

```python
# 2. 앞에서 구한 데이터에 'views' 컬럼 결측치를 'views' 컬럼의 최솟값으로 채워주세요.
min = df['views'].min() # 42.0
df['views'] = df['views'].fillna(min)
# df[df['views'].isnull()] # 결측치가 하나도 없으므로, 최솟값으로 잘 채워진 것임.
df
```

```python
# 3. 그리고 'views' 컬럼의 중앙값을 계산해 정수로 구하시오.
mid = df['views'].median() # 5924.0
print(int(mid))
```
